```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('LST_day.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

## LinePlot with Hue

![image](lineplot.jpg)

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='date', y='mean', hue='KGISWardID')
```

```python
mean_per_ward = df.groupby('KGISWardID')['mean']
```

```python
max_mean_per_word = mean_per_ward.max()
```

```python
top_3_wards = max_mean_per_word.nlargest(5).index
```

```python
filtered_df = df[df['KGISWardID'].isin(top_3_wards)]
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.lineplot(data=filtered_df, x='date', y='mean', hue='KGISWardNa')
```
